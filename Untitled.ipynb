## Opening Price of Stock

###  Graph Of Opening Price Stock

```python
plt.figure(figsize=(20,9))
plt.plot(ds.Open)
plt.xlabel("Day")
plt.ylabel("Price of Stock")
plt.grid(axis= "y")
plt.title("Opening Price of Stock")
plt.show()
```

###  Visulaization Of Moving Average

```python
md50=ds.Open.rolling(50).mean()
```

```python
md100=ds.Open.rolling(100).mean()
```

```python
plt.figure(figsize=(20,15))
plt.plot(ds.Open)
plt.plot(md50, "r")
plt.plot(md100, "g")
plt.title("Moving Avg of 50 & 100 Days")
```

### Model  Preprocessing

```python
ds["fo"] = ds[['Open']].shift(-1)
```

```python
ds.tail(10)
```

```python
df_o = ds[['Open','fo']]
```

```python
df_o.tail(1)
```

### Spliting Data

```python
X = np.array(df_o.drop(['fo'],1))[:-1]
```

```python
y = np.array(df_o['fo'])[:-1]
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=110)
```

### Fitting Model

#### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
dtr= DecisionTreeRegressor()
dtr.fit(X_train,y_train)
```

####  Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rr = RandomForestRegressor()
rr.fit(X_train,y_train)
```

## Predictng Model

```python
x_p = df_o['Open'][:-1]
x_p = x_p.tail(1)
x_p = np.array(x_p).reshape(-1,1)
```

```python
l_p = np.array(15440.0).reshape(-1,1)
```

```python
y_pred = dtr.predict(l_p)
print(y_pred)
```

```python
y_pred1 = lr.predict(l_p)
print(y_pred1)
```

```python
y_pred2 = rr.predict(l_p)
print(y_pred2)
```
